思考一下我們前面有提到圖片是矩陣，但維度可能會不一樣
例如灰階圖只有兩個維度，RGB 彩圖則有 3 個維度

假如今天我們把 RGB 3 個維度拆開來看會有甚麼不同的效果呢？

```python
from matplotlib import pyplot as plt
import cv2
import numpy as np
```

```python
img_path = 'data/lena.png'

img = cv2.imread(img_path, cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
# R
img_r = img[:,:,0]
# G
img_g = img[:,:,1]
# B
img_b = img[:,:,2]
```

```python
plt.subplot(121)
plt.imshow(img_r, cmap='Reds')
plt.subplot(122)
plt.imshow(img_r, cmap=plt.cm.gray)
plt.show()
```

```python
plt.subplot(121)
plt.imshow(img_g, cmap='Greens')
plt.subplot(122)
plt.imshow(img_g, cmap=plt.cm.gray)
plt.show()
```

```python
plt.subplot(121)
plt.imshow(img_b, cmap='Blues')
plt.subplot(122)
plt.imshow(img_b, cmap=plt.cm.gray)
plt.show()

plt.show()
```

```python
(R,G,B) = cv2.split(img)
```

```python
while True:
    cv2.imshow('R', R)
    cv2.imshow('G', G)
    cv2.imshow('B', B)
    k = cv2.waitKey(0)
    if k == 27:
        cv2.destroyAllWindows()
        break
```
